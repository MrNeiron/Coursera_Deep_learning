```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def PadZero(x, pad = 2):
    xPad = np.pad(x,((0,0),(pad,pad),(pad,pad),(0,0)),"constant",constant_values = 0)

    return xPad
```

```python
pad = 2
m = 2 #Number of examples
nC = 3 #Number of chanels
nH = 8 #Height of image
nW = 8 #Width of image
x = np.random.randint(0,256,(m,nH,nW,nC))
xPad = PadZero(x, pad)
print("{}->{}\n{}\n".format(x.shape,xPad.shape,xPad[0,:,:,0]))

fig1, axarr1 = plt.subplots(1, 2)
axarr1[0].set_title('x1{}'.format(x.shape))
axarr1[0].imshow(x[0,:,:,0])
axarr1[1].set_title('xPad1{}'.format(xPad.shape))
axarr1[1].imshow(xPad[0,:,:,0])

fig2, axarr2 = plt.subplots(1, 2)
axarr2[0].set_title('x2{}'.format(x.shape))
axarr2[0].imshow(x[0,:,:,1])
axarr2[1].set_title('xPad2{}'.format(xPad.shape))
axarr2[1].imshow(xPad[0,:,:,1])

fig3, axarr3 = plt.subplots(1, 2)
axarr3[0].set_title('x3{}'.format(x.shape))
axarr3[0].imshow(x[0,:,:,2])
axarr3[1].set_title('xPad3{}'.format(xPad.shape))
axarr3[1].imshow(xPad[0,:,:,2])
    
    
```
